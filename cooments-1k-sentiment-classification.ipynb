```python
import pandas as pd
from textblob import TextBlob

def analyze_sentiment(text):
    analysis = TextBlob(text)
    if analysis.sentiment.polarity >= 0:
        return 'pos'
    else:
        return 'neg'

# 读取文件内容并跳过第一行
file_path = 'comments-1k.txt'
with open(file_path, 'r', encoding='utf-8') as file:
    lines = file.readlines()[1:]

# 假设文件中每一行是一个评论，格式为 "评论内容"
data = []
for line in lines:
    line = line.strip()
    if line:
        data.append(line)

# 转换为DataFrame
df = pd.DataFrame(data, columns=['text'])

# 预测情感
df['predicted_label'] = df['text'].apply(analyze_sentiment)

# 打印前几行数据检查预测结果
df.head(1000)
```

```python
import pandas as pd
from sklearn.metrics import classification_report
from textblob import TextBlob

def analyze_sentiment(text):
    analysis = TextBlob(text)
    if analysis.sentiment.polarity >= 0:
        return 'pos'  # pos代表积极
    else:
        return 'neg'  # neg代表消极

# 读取CSV文件
file_path = 'comments-1k.txt'
df = pd.read_csv(file_path)

# 将 '0' 转换为 'neg'，'1' 转换为 'pos'
df['label'] = df['label'].map({0: 'neg', 1: 'pos'})

# 预测情感
df['predicted_label'] = df['review'].apply(analyze_sentiment)

# 计算准确率、召回率和F1值
report = classification_report(df['label'], df['predicted_label'], output_dict=True)

# 打印结果
report_df = pd.DataFrame(report).transpose()
report_df
```
